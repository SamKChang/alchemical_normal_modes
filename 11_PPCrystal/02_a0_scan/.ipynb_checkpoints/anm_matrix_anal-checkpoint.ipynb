# Gradiant vector and Hessian matrix construction

```python
def get_dicts():
    H_dict = {}
    G_dict = {}
    a_ref_dict = {}
    qmouts = {}
    for info_cyl in cyl_map:
        base = info_cyl[0]
        H_root = '%s_H_scan' % base
        delta = 0.01
        H_list = []
        G_list = []
        outs = []
        a_ref_list = []
        H_dict[base] = H_list
        G_dict[base] = G_list
        a_ref_dict[base] = a_ref_list
        qmouts[base] = outs
        for info_a in cyl_map:
            tar = info_a[0]
            N = len(sorted(glob('%s/%s_a-%s_i??[pm]-I*' % (H_root, base, tar)))) / 2
            H = np.zeros([N, N])
            G = np.zeros(N)
            H_list.append(H)
            G_list.append(G)

            a_ref_file = glob("%s/%s_a-%s/*.out" % (H_root, base, tar))[0]
            a_ref = qtk.QMOut(a_ref_file, program='cpmd')
            a_ref_list.append(a_ref)
            outs.append(a_ref)

            di_p = []
            di_m = []
            for i in range(N):
                di_p_file = glob('%s/%s_a-%s_i%02dp-I*/*.out' % (H_root, base, tar, i))[0]
                di_m_file = glob('%s/%s_a-%s_i%02dm-I*/*.out' % (H_root, base, tar, i))[0]
        #         di_p_file = glob('%s/%s_a-%s_i%03dp-I*/*.out' % (H_root, base, tar, i))[0]
        #         di_m_file = glob('%s/%s_a-%s_i%03dm-I*/*.out' % (H_root, base, tar, i))[0]
                dp = qtk.QMOut(di_p_file, program='cpmd')
                dm = qtk.QMOut(di_m_file, program='cpmd')
                di_p.append(dp)
                di_m.append(dm)
                outs.append(dp)
                outs.append(dm)
                H[i, i] = (dp + dm - 2*a_ref) / (delta)**2
                G[i] = (dp - dm) / (2*delta)

            for i in range(N):
                for j in range(i+1, N):
                    didj_file = glob('%s/%s_a-%s_i%02dj%02d*/*.out' % (H_root, base, tar, i, j))[0]
                    #didj_file = glob('%s/%s_a-%s_i%03dj%03d*/*.out' % (H_root, base, tar, i, j))[0]
                    didj = qtk.QMOut(didj_file, program='cpmd')
                    outs.append(didj)

                    H[i, j] = (((didj - di_p[j]) / delta) - ((di_p[i] - a_ref) / delta)) / delta
                    H[j, i] = H[i, j]        

        print len(H_list)
    return H_dict, G_dict, a_ref_dict
```

# Prediction test

```python
ref_dir = 'ref'
outs_all = []
for info_a in cyl_map:
    outs = [info_a[0], []]
    outs_all.append(outs)
    for info_cyl in cyl_map:
        cyl = info_cyl[0]
        tar = info_a[0]
        out_file = glob('%s/%s_a-%s/*.out' % (ref_dir, cyl, tar))[0]
        out = qtk.QMOut(out_file, program='cpmd')
        #outs[1].append([out.name.split('.')[0], out])
        outs[1].append(out)
```

```python
mol_base = qtk.Molecule('xyz/ge.xyz')
```

```python
def cyl2par(mol, size=[1, 3]):
    mol.sort(order='xyz')
    par = []
    for a in mol.type_list:
        crd = qtk.PP(a, size=size)
        vec = crd.vectorize()
        par.append([vec[0][1:], a, vec[0][0]])
    return par

def par2cyl(par, mol_base=mol_base, size=[1, 3]):
    mol = mol_base.copy()
    mol.sort(order='xyz')
    for i, crd in enumerate(par):
        print crd
        pp = qtk.PP(crd[1], size=size)
        vec = [crd[2]]
        vec.extend(crd[0])
        pp.unvectorize(vec, size[0], range(size[1], 0, -1))
        pp.name = crd[1] + '%02d' % i
        mol.setAtoms(i, string=pp)
    return mol

def mol2vec(mol_tar, mol_ref, size=[1, 3]):
    par_tar = cyl2par(mol_tar)
    par_ref = cyl2par(mol_ref)
    vec_tar = np.hstack([p[0] for p in par_tar])
    vec_ref = np.hstack([p[0] for p in par_ref])
    return vec_tar - vec_ref
```

```python
tars_dict = {}
prd_1_dict = {}
prd_2_dict = {}
crd_dict = {}

for cyl, G_list in G_dict.iteritems():
    tars = []
    prd_1 = []
    prd_2 = []
    crds = defaultdict(list)
    tars_dict[cyl] = tars
    prd_1_dict[cyl] = prd_1
    prd_2_dict[cyl] = prd_2
    crd_dict[cyl] = crds
    for a_pos in range(len(a_ref_dict[cyl])):
        ref = a_ref_dict[cyl][a_pos]
        for t, tar in enumerate(outs_all[a_pos][1]):
            vec = mol2vec(tar.molecule, ref.molecule)
            crds[cyl_map[t][0]].append(vec)
            try:
                dE1 = G_dict[cyl][a_pos].dot(vec)
                dE2 = vec.dot(H_dict[cyl][a_pos].dot(vec)) / 2
                # only compute predictions to other crystals
                if sum(abs(vec)) > 0:
                    tars.append(tar.Et)
                    prd_1.append(ref.Et + dE1)
                    prd_2.append(ref.Et + dE1 + dE2)
            except ValueError:
                print "failed: ", cyl, i, G_dict[cyl][a_pos].shape
```

```python
def get_cyl(ax, crd, colors=['C0', 'k'], width=.8, depth=.2, size=.5, radius=.22, fraction=.55):
    
    signs = [1, -1]
    A = crd
    C = tuple(a + size for a in A)
    F = tuple(c + size * depth for c in C)
    B = tuple((C[i] + A[i]) / 2. - signs[i] * width / 2. * size for i in range(2))
    D = tuple((C[i] + A[i]) / 2. + signs[i] * width / 2. * size for i in range(2))
    E = tuple(b + size * depth for b in B)
    G = tuple(d + size * depth for d in D)
    H = tuple(C[i] * fraction + A[i] * (1-fraction) for i in range(2))
    
    # crystal box
    ax.add_patch(Polygon([A, B, C, D], closed=True, fill=False, hatch='', lw=.2))
    ax.add_patch(Polygon([B, E, F, C], closed=True, fill=False, hatch='', lw=.2))
    ax.add_patch(Polygon([D, G, F, C], closed=True, fill=False, hatch='', lw=.2))
    ax.add_patch(Circle(A, radius * size, facecolor=colors[0], edgecolor='k'))
    ax.add_patch(Circle(H, radius * size, facecolor=colors[1], edgecolor='k'))

def crystal_plot(ind, ax=None):
    
    if ax is None:
        fig, ax = plt.subplots()
    ax.set_aspect('equal')
    
    Zs = cyl_map[ind][2]
    
    cs = 2.3
    get_cyl(ax, (cs, .2), get_colors(Zs))
    
    Zs_list = decomp(ind)
    
    if len(Zs_list) == 2:
        
        get_cyl(ax, (cs + 1.5, .2), get_colors(Zs_list[0]))
        get_cyl(ax, (cs + 2.8, .2), get_colors(Zs_list[1]))
        ax.text(cs + .7, 0.45, '=', fontsize=20, verticalalignment='center')
        ax.text(cs + 2.1, 0.45, '+', fontsize=20, verticalalignment='center')
        
#         ax.text(cs + 3.5, 0.45, '=', fontsize=20, verticalalignment='center')
#         ax.text(cs + 4.2, 0.45, Zs2N(Zs_list[0]), fontsize=15, verticalalignment='center', horizontalalignment='left')
#         ax.text(cs + 6, 0.45, '+', fontsize=20, verticalalignment='center')
#         ax.text(cs + 6.7, 0.45, Zs2N(Zs_list[1]), fontsize=15, verticalalignment='center', horizontalalignment='left')
    
    ax.plot([cs - .7], [.5], marker=filled_markers[ind], markerfacecolor="None", 
            markersize=10, color='k', markeredgewidth=1.5)
    ax.text(cs - 1.2, 0.45, Zs2N(Zs), fontsize=15, verticalalignment='center', horizontalalignment='right')
    
    ax.axis('off')
    ax.set_xlim(0, 6)
    ax.set_ylim(0, 1)
```

```python
def periodic_table(ax):
    Zs = [13, 14, 15, 31, 32, 33, 49, 50, 51]
    arr = [[], [], []]
    for i, Z in enumerate(Zs):
        row, col = divmod(i, len(arr))
        rgb = mpl.colors.to_rgb(color_dict[qtk.Z2n(Z)])[:3]
        arr[row].append(rgb)
        txt = ax.text(col, row, qtk.Z2n(Z), verticalalignment='center', horizontalalignment='center', fontsize=15)
        txt.set_path_effects([PathEffects.withStroke(linewidth=5, foreground='w')])
    arr = np.asarray(arr)
    ax.imshow(arr)
    ax.set_xticks([], [])
    ax.set_yticks([], [])
    
```
