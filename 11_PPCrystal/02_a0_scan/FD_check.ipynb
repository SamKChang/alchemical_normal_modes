```python
import qctoolkit as qtk
import numpy as np
from glob import glob
import os
```

```python
files = sorted(glob('ge_H_scan/*/*.out'))
new_files = sorted(glob('ge_H_scan_new/*/*.out'))
```

```python
outs, new_outs = [], []
for i, f1 in enumerate(files):
    f2 = new_files[i]
    outs.append(qtk.QMOut(f1, program='cpmd'))
    new_outs.append(qtk.QMOut(f2, program='cpmd'))
```

```python
for i, o1 in enumerate(outs):
    o2 = new_outs[i]
    print o1 - o2
```
