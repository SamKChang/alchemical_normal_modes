```python
mol_base = qtk.Molecule('coronene.xyz')
mol_base.view2D()
ccs = qtk.CCS(mol_base, 'coronene.yml')
dZ_test = np.array(ccs.random_coord()['mutation'][0]) - 6
```

```python
def dZ2mol(dZ, mol_base=mol_base, get_index=False):
    
    dZ = np.asarray(dZ)
    mol_base.Z = np.asarray(mol_base.Z)
    
    Z = np.ones(len(mol_base.Z))
    Z[:len(dZ)] = mol_base.Z[:len(dZ)] + dZ
    R = mol_base.R
    ZR = np.hstack([Z[:,np.newaxis], R])
    
    mol = qtk.Molecule()
    mol.build(ZR)
    #mol.center()
    return mol

def rotate(dZ_in, angle=0, show=False, flip=False, get_index=False):
    
    mol0 = dZ2mol(dZ_in)
    mol0.center(mol0.getCenterOfMass())
    
    mol = mol0.copy()
    mol.center(mol.getCenterOfMass())
    
    if flip:
        mol.rotate(np.pi, u=[1,0,0])
    if angle:
        mol.rotate(angle, u=[0,0,1])
    
    if show:
        display(SVG(qtk.rdk.mol2svg(mol)))
        
    dZ = np.ones(mol.element_count())
    ind = range(len(dZ))
    
    for i in range(mol.element_count()):
        for j in range(mol.element_count()):
            if np.linalg.norm(mol.R[i] - mol0.R[j]) < 1.:
                dZ[j] = mol.Z[i] - 6
                ind[j] = i
    if get_index:
        return dZ, ind
    else:
        return dZ
```

# Counting duplicates

```python
def gen_12_dZs(dZ_base, get_index=False):
    """Used to generate all 12 rotations and flips
       the same operation on index is the same all other molecules
    """
    dZs = []
    inds = []
    angles = np.arange(0, 2 * np.pi, np.pi / 3)
    for a in angles:
        dZ, ind = rotate(dZ_base, a, get_index=True)
        dZs.append(dZ)
        inds.append(ind)
    for a in angles:
        dZ, ind = rotate(dZ_base, a, flip=True, get_index=True)
        dZs.append(dZ)
        inds.append(ind)
    if get_index:
        return np.vstack(dZs), np.vstack(inds)
    else:
        return np.vstack(dZs)
    
def gen_12_dZs_inds(dZ_base, inds):
    """use generated list of roation indices to generate roated dZs"""
    dZs = np.zeros([len(inds), len(dZ_base)])
    for i, ind in enumerate(inds):
        dZs[i] = dZ_base[ind]
    return dZs

def count_uniq_dZs(dZs_in):
    """count unique rows in dZs_in"""
    data = dZs_in

    sorted_idx = np.lexsort(data.T)
    sorted_data =  data[sorted_idx,:]
    row_mask = np.append([True],np.any(np.diff(sorted_data,axis=0),1))
    out = sorted_data[row_mask]
    return len(out)
```

# load all 1.8m coronene data and compute duplicates

```python
mol_data_files = sorted(glob('../02_ANM_modePlot/data_space/*-dz*.txt'))

mol_data_cr = []
for f in mol_data_files:
    mol_data_cr.append(np.loadtxt(f))

E_data_files = sorted(glob('../02_ANM_modePlot/data_space/*-dE*.txt'))
E_data_grp = []
for f in E_data_files:
    E_data_grp.append(np.loadtxt(f))
```

```python
dup_grp = []
for dZs in mol_data_cr:
    dups = []
    dup_grp.append(dups)
    for dZ in dZs:
        dZs_rot = gen_12_dZs_inds(dZ, inds)
        dups.append(count_uniq_dZs(dZs_rot))
```

```python
def data_group(ind):
    Delta_E = E_data_grp[ind][:,0]
    E_nn = E_data_grp[ind][:,1]
    Delta_Z = mol_data_cr[ind].astype(int)
    dup = dup_grp[ind]
    return Delta_E, E_nn, Delta_Z, dup

def df_append(df_in, i):
    df = pd.DataFrame()
    dE, En, dZ, dup = data_group(i)
    N = np.ones(len(En)).astype(int) * (i + 1)
    df['Delta_E'] = dE
    df['Enn'] = En
    df['Delta_Z'] = pd.Series([dz.tolist() for dz in dZ])
    df['N_BN'] = N
    df['N_dup'] = dup
    return df_in.append(df)
    
df_dup = pd.DataFrame()
for i in range(len(inds)):
    df_dup = df_append(df_dup, i)
```
