```python
##########################################################################
# convert molecular structure in Bohr into PDB format for RDKit plotting #
##########################################################################
# once required *.pdb files is generated
# this procedure is no longer necessary

def data2mol(data, name, align_index=None):
    """convert npz molecule data to xyz"""
    Z = data['Z0'][:, np.newaxis]
    R = data['coor'] / qtk.setting.a2b
    ZR = np.hstack([Z, R])
    mol = qtk.Molecule()
    mol.build(ZR)
    mol.name = name
    if align_index:
        mol.alignAtoms(*align_index)
    else:
        mol.align()
    return mol
```

```python
from rdkit import Chem
from rdkit.Chem import AllChem

from rdkit.Chem.Draw import IPythonConsole #Needed to show molecules
from rdkit.Chem.Draw.MolDrawing import MolDrawing, DrawingOptions #Only needed if modifying defaults

from rdkit.Chem import rdDepictor
from rdkit.Chem.Draw import rdMolDraw2D
from IPython.display import SVG, display, HTML
```

```python
def mol2svg(mol,
             molSize=(200,200),
             kekulize=False, 
             index=False,
             atom_label=False,
             highlight=[],
             colors={},
             sizes={},
            ):
    
    #drawer = rdMolDraw2D.MolDraw2DSVG(molSize[0],molSize[1], int(.5*molSize[0]), int(molSize[1]*.5))
    drawer = rdMolDraw2D.MolDraw2DSVG(molSize[0],molSize[1])
    opts = drawer.drawOptions()
    kw_drawing = {}
    
    ##################
    # bare bone plot #
    ##################
    m = Chem.Mol(mol.ToBinary())
    if kekulize:
        try:
            Chem.Kekulize(m)
        except:
            m = Chem.Mol(mol.ToBinary())
    if not m.GetNumConformers():
        rdDepictor.Compute2DCoords(m)
    
    ############
    # indexing #
    ############
    if index:
        for i in range(m.GetNumAtoms()):
            opts.atomLabels[i] = m.GetAtomWithIdx(i).GetSymbol()+str(i)
            
    ##############
    # atom label #
    ##############
    if atom_label:
        for i in range(m.GetNumAtoms()):
            opts.atomLabels[i] = m.GetAtomWithIdx(i).GetSymbol()
        
    #################
    # high lighting #
    #################
    kw_hl = {
        'highlightAtoms': highlight,
        'highlightAtomColors': colors,
        'highlightAtomRadii': sizes,
        'highlightBonds': None,
    }
    kw_drawing.update(kw_hl)
    
    drawer.DrawMolecule(m, **kw_drawing)
    drawer.FinishDrawing()
    svg = drawer.GetDrawingText()
    return svg.replace('svg:','')
```

```python
def rdmol2svglist(m, data):
    svg_list = []
    for i in range(len(data['w'])):
        v = data['v'][:,i]
        w = data['w'][i]
        color = {}
        sizes = {}
        highlight = []
        for j in range(len(data['w'])):
            highlight.append(j)
            sizes[j] = v[j]
            if v[j] != 0:
                if v[j] > 0:
                    color[j] = tuple(192./255 for _ in range(3))
                elif v[j] < 0:
                    color[j] = tuple(.2 for _ in range(3))
        kw_bz1 = {
            #'molSize': (400,400),
            'molSize': (150, 150),
            'index': False,
            # highlights
            'highlight': highlight,
            'colors': color,
            'sizes': sizes
        }
        svg_list.append(mol2svg(m, **kw_bz1))
    return svg_list
```

```python
def svg2file(svgs, name):
    for i, s in enumerate(svgs):
        with open('%s_ev_%02d.svg' % (name, i), 'w') as f:
            f.write(s)
```

```python
def energy_estimate(dZ):
    # normal mode decomposition
    dZes = dZ.dot(data_cr['v'])
    # calculate first/second order estimate
    d1E = dZ.dot(d1E_Z)
    d2E = (dZes ** 2).dot(data_cr['v']) * 0.5
    # wired shape problem
    return d1E + d2E.sum(1)
```
